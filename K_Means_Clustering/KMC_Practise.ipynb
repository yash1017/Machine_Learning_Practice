```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_blobs
```

```python
make_blobs
```

```python
data = make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=2)
```

```python
data[0]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig,axis = plt.subplots(1,2,)
axis[0].scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')

axis[1].scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```
